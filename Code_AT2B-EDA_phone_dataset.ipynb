# 2.0 Filling Missing values

```python
data.isnull().sum()
```

```python
columns_to_fill = [
  'rear_camera_mp', 'front_camera_mp', 'internal_memory',
  'ram', 'battery', 'weight'
]
data[columns_to_fill] = data.groupby('device_brand')[columns_to_fill].transform(
  lambda x: x.fillna(x.median())
)
```

```python
data['device_brand'].unique()
```

```python
data.isnull().sum()
```

```python
print(data.groupby('device_brand')['rear_camera_mp'].median())
```

```python
overall_median = data['rear_camera_mp'].median()
data['rear_camera_mp'] = data.groupby('device_brand')['rear_camera_mp'].transform(
  lambda x: x.fillna(x.median())
)
data['rear_camera_mp'].fillna(overall_median, inplace=True)
```

```python
data.isnull().sum()
```

# 3.0 EDA

## 3.1 Correlation Map

```python
corr_matrix = data.drop(columns=['normalized_new_price']).corr(numeric_only=True)

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

Should drop weight, battery, and screen size as they highly corelate with each other: *multicollinearity issue*

```python
plt.figure(figsize = (20,15))
sns.countplot(data=data, x='device_brand', order=data['device_brand'].value_counts().index, palette='Set2')
plt.xticks(rotation=45)
plt.title('Device Brand Count')
plt.xlabel('Device Brand')
plt.ylabel('Count')
plt.show()
```

```python
top_10_brands = data['device_brand'].value_counts().nlargest(10).index

plt.figure(figsize=(20, 15))
sns.countplot(data=data[data['device_brand'].isin(top_10_brands)],x='device_brand',order=top_10_brands,palette='Set2')

plt.xticks(rotation=45)
plt.title('Top 10 Device Brands by Count')
plt.xlabel('Device Brand')
plt.ylabel('Count')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 4))

sns.countplot(data=data, x='os', ax=axes[0,0], palette='Set2')
sns.countplot(data=data, x='4g', ax=axes[0,1], palette='Set2')
sns.countplot(data=data, x='5g', ax=axes[1,0], palette='Set2')
sns.countplot(data=data, x='release_year', ax=axes[1,1], palette='Set2')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(7, figsize=(10, 20))

sns.histplot(data['normalized_used_price'],ax=axes[0] )
sns.histplot(data['rear_camera_mp'], ax=axes[1])
sns.histplot(data['front_camera_mp'], ax=axes[2])
sns.histplot(data['internal_memory'], ax=axes[3])
sns.histplot(data['ram'], ax=axes[4])
sns.histplot(data['battery'], ax=axes[5])
sns.histplot(data['weight'], ax=axes[6])

plt.tight_layout()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['normalized_used_price'])
plt.title('Histogram of normalized_used_price')
plt.xlabel('normalized_used_price')
plt.ylabel('Frequency')
plt.show()
```

## 3.2 Depreciation Percentage Graph

```python
data['depreciation_percentage'] = (data['normalized_used_price'] / data['normalized_new_price']) * 100
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['depreciation_percentage'], bins=30, kde=True)
plt.title('Distribution of Depreciation Percentage')
plt.xlabel('Depreciation Percentage (%)')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
top_10_brands = data['device_brand'].value_counts().nlargest(10).index
sns.boxplot(data=data[data['device_brand'].isin(top_10_brands)],
  x='device_brand',
  y='depreciation_percentage',
  palette='Set3')
plt.xticks(rotation=45)
plt.title('Depreciation Percentage by Top 10 Brands')
plt.ylabel('Depreciation Percentage (%)')
plt.xlabel('Device Brand')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x='os', y='depreciation_percentage', palette='Set2')
plt.title('Depreciation Percentage by Operating System')
plt.ylabel('Depreciation Percentage (%)')
plt.xlabel('OS')
plt.grid(True)
plt.show()
```
